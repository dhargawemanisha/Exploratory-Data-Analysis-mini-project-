#  Find correlation between variables

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

#  Plot different features

```python
# Pairplot to visualize relationships between numerical features
sns.pairplot(df, hue='species')
plt.show()

# Box plot to visualize the distribution of a numerical feature based on a categorical variable
sns.boxplot(x='species', y='culmen_length_mm', data=df)
plt.title('Culmen Length by Species')
plt.show()

# Histogram to visualize the distribution of a numerical feature
sns.histplot(data=df, x='body_mass_g', bins=20, kde=True)
plt.title('Distribution of Body Mass')
plt.show()
```

# Detect outlier / missing values

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns='species'))
plt.title('Box Plot for Outlier Detection')
plt.xticks(rotation=45)
plt.show()
```
